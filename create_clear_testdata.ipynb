```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
import random
```

```python
test_filepath = 'data/target_test.csv'
testdata = pd.read_csv(test_filepath)
```

```python
testdata
```

```python
reference_filepath = 'data/target_train.csv'
reference = pd.read_csv(reference_filepath)
```

```python
ASIaverage = pd.pivot_table(reference, index = ["Anon Student Id"])
ASIDURA = ASIaverage["Step Duration (sec)"]
PNaverage = pd.pivot_table(reference, index = ["Problem Name"])
PNDURA = PNaverage["Step Duration (sec)"]
STEPaverage = pd.pivot_table(reference, index = ["Step Name"])
STEPDURA = STEPaverage["Step Duration (sec)"]
PHaverage = pd.pivot_table(reference, index = ["Problem Hierarchy"])
PHDURA = PHaverage["Step Duration (sec)"]
```

```python
tlist = [i for i in range(1140)]
```

```python
newdata = pd.DataFrame(columns = ["A"], index = tlist)
```

```python
newdata.insert(0, "Anon Student Id", testdata["Anon Student Id"])
del newdata["A"]
```

```python
dealmean = STEPDURA.mean()
```

```python
newdata.insert(1, "Problem Hierarchy", testdata["Problem Hierarchy"])
newdata.insert(2, "Problem Name", testdata["Problem Name"])
newdata.insert(3, "Problem View", testdata["Problem View"])
newdata.insert(4, "Step Name", testdata["Step Name"])
newdata.insert(5, "KC(Default)", testdata["KC(Default)"])
newdata.insert(6, "Opportunity(Default)", testdata["Opportunity(Default)"])
```

```python
newdata.insert(7, "Step Duration (Anon Student Id)", np.ones(1140))
newdata.insert(8, "Step Duration (Problem Name)", np.ones(1140))
newdata.insert(9, "Step Duration (Step Name)", np.ones(1140))
newdata.insert(10, "Step Duration (Problem Hierarchy)", np.ones(1140))
```

```python
for i in range(1140):
    newdata["Step Duration (Anon Student Id)"][i] = ASIDURA[testdata["Anon Student Id"][i]]
    newdata["Step Duration (Problem Name)"][i] = PNDURA[testdata["Problem Name"][i]]
    newdata["Step Duration (Problem Hierarchy)"][i] = PHDURA[testdata["Problem Hierarchy"][i]]
    if pd.isnull(newdata["Step Name"][i]):
        newdata["Step Duration (Step Name)"][i] = dealmean
    else:
        newdata["Step Duration (Step Name)"][i] = STEPDURA[testdata["Step Name"][i]]
    if pd.isnull(newdata["Opportunity(Default)"][i]):
        newdata["Opportunity(Default)"][i] = 0
    else:
        value = newdata["Opportunity(Default)"][i].split("~~")
        initial = 0
        for k in value:
            initial = initial + int(k)
        newdata["Opportunity(Default)"][i] = initial
```

```python
newdata.insert(0, "Row", testdata["Row"])
newdata.insert(12, "Correct First Attempt", testdata["Correct First Attempt"])
```
