```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
import random
```

```python
train_filepath = 'data/train.csv'
traindata = pd.read_table(train_filepath)
```

```python
ASIaverage = pd.pivot_table(traindata, index = ["Anon Student Id"])
ASICFAR = ASIaverage["Correct First Attempt"]
ASIDURA = ASIaverage["Step Duration (sec)"]
```

```python
PNaverage = pd.pivot_table(traindata, index = ["Problem Name"])
PNCFAR = PNaverage["Correct First Attempt"]
PNDURA = PNaverage["Step Duration (sec)"]
```

```python
STEPaverage = pd.pivot_table(traindata, index = ["Step Name"])
STEPCFAR = STEPaverage["Correct First Attempt"]
STEPDURA = STEPaverage["Step Duration (sec)"]
```

```python
PHaverage = pd.pivot_table(traindata, index = ["Problem Hierarchy"])
PHCFAR = PHaverage["Correct First Attempt"]
PHDURA = PHaverage["Step Duration (sec)"]
```

```python
tlist = [i for i in range(232744)]
```

```python
newdata = pd.DataFrame(columns = ["A"], index = tlist)
```

```python
newdata.insert(0, "Anon Student Id", traindata["Anon Student Id"])
```

```python
newdata.insert(1, "Problem Hierarchy", traindata["Problem Hierarchy"])
```

```python
newdata.insert(2, "Problem Name", traindata["Problem Name"])
```

```python
newdata.insert(3, "Problem View", traindata["Problem View"])
```

```python
newdata.insert(4, "Step Name", traindata["Step Name"])
```

```python
newdata.insert(5, "KC(Default)", traindata["KC(Default)"])
```

```python
newdata.insert(6, "Opportunity(Default)", traindata["Opportunity(Default)"])
```

```python
newdata.insert(7, "CFAR (Anon Student Id)", np.ones(232744))
newdata.insert(8, "CFAR (Problem Name)", np.ones(232744))
newdata.insert(9, "CFAR (Step Name)", np.ones(232744))
newdata.insert(10, "CFAR (Problem Hierarchy)", np.ones(232744))
newdata.insert(11, "Step Duration (Anon Student Id)", np.ones(232744))
newdata.insert(12, "Step Duration (Problem Name)", np.ones(232744))
newdata.insert(13, "Step Duration (Step Name)", np.ones(232744))
newdata.insert(14, "Step Duration (Problem Hierarchy)", np.ones(232744))
```

```python
newdata.insert(7, "Step Duration (sec)", traindata["Step Duration (sec)"])
```

```python
tnan = newdata["KC(Default)"][0]
for i in range(232744):
    newdata["CFAR (Anon Student Id)"][i] = ASICFAR[traindata["Anon Student Id"][i]]
    newdata["CFAR (Problem Name)"][i] = PNCFAR[traindata["Problem Name"][i]]
    newdata["CFAR (Step Name)"][i] = STEPCFAR[traindata["Step Name"][i]]
    newdata["CFAR (Problem Hierarchy)"][i] = PHCFAR[traindata["Problem Hierarchy"][i]]
    newdata["Step Duration (Anon Student Id)"][i] = ASIDURA[traindata["Anon Student Id"][i]]
    newdata["Step Duration (Problem Name)"][i] = PNDURA[traindata["Problem Name"][i]]
    newdata["Step Duration (Step Name)"][i] = STEPDURA[traindata["Step Name"][i]]
    newdata["Step Duration (Problem Hierarchy)"][i] = PHDURA[traindata["Problem Hierarchy"][i]]
    if newdata["Step Duration (sec)"][i] == tnan:
        newdata["Step Duration (sec)"][i] = (newdata["CFAR (Anon Student Id)"][i] + newdata["CFAR (Problem Name)"][i] + newdata["CFAR (Step Name)"][i])/3
```

```python
dealmean = STEPDURA.mean()
```

```python
STEPDURA = STEPDURA.fillna(dealmean)
len(STEPDURA.dropna())
```

```python
for i in range(232744):
    if pd.isnull(newdata["Step Duration (Step Name)"][i]):
        newdata["Step Duration (Step Name)"][i] = STEPDURA[traindata["Step Name"][i]]
    if pd.isnull(newdata["Step Duration (sec)"][i]):
        newdata["Step Duration (sec)"][i] = (newdata["CFAR (Anon Student Id)"][i] + newdata["CFAR (Problem Name)"][i] + newdata["CFAR (Step Name)"][i])/3
    if pd.notnull(newdata["Opportunity(Default)"][i]):
        value = newdata["Opportunity(Default)"][i].split("~~")
        initial = 0
        for k in value:
            initial = initial + int(k)
        newdata["Opportunity(Default)"][i] = initial
```
