```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
import random
```

```python
import pyspark as ps
conf = ps.SparkConf()
sc = ps.SparkContext(conf = conf)
```

```python
traindata = pd.read_table("data/train.csv")
trainstep = pd.pivot_table(traindata, index = ["Step Name"])
stepcfar = trainstep["Correct First Attempt"]
```

```python
dealmean = stepcfar.mean()
```

```python
testdata = pd.read_csv("data/target_clear_test.csv")
```

```python
numlist = [i for i in range(1140)]
```

```python
numrdd = sc.parallelize(numlist)
```

```python
def deal(x):
    if pd.isnull(testdata["Step Name"][x]):
        testdata["Step Name"][x] = dealmean
    return testdata["Step Name"][x]

kd = numrdd.map(deal)
kd.collect()
```

```python
testdata["Step Name"]
```

```python
del testdata["Step Name"]
testdata.insert(6, "Step Name", kd.collect())
```
