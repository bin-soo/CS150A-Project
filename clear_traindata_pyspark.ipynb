```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
import random
```

```python
train_filepath = 'data/train.csv'
traindata = pd.read_table(train_filepath)
```

```python
ASIaverage = pd.pivot_table(traindata, index = ["Anon Student Id"])
ASIDURA = ASIaverage["Step Duration (sec)"]
PNaverage = pd.pivot_table(traindata, index = ["Problem Name"])
PNDURA = PNaverage["Step Duration (sec)"]
STEPaverage = pd.pivot_table(traindata, index = ["Step Name"])
STEPDURA = STEPaverage["Step Duration (sec)"]
PHaverage = pd.pivot_table(traindata, index = ["Problem Hierarchy"])
PHDURA = PHaverage["Step Duration (sec)"]

dealmean = STEPDURA.dropna().mean()
STEPDURA = STEPDURA.fillna(dealmean)
```

```python
import pyspark as ps
conf = ps.SparkConf()
sc = ps.SparkContext(conf = conf)
```

```python
asilist = traindata["Anon Student Id"]
pnlist = traindata["Problem Name"]
steplist = traindata["Step Name"]
phlist = traindata["Problem Hierarchy"]

asirdd = sc.parallelize(asilist)
pnrdd = sc.parallelize(pnlist)
steprdd = sc.parallelize(steplist)
phrdd = sc.parallelize(phlist)
```

```python
def find(x):
    return ASIDURA[x]

asiduralist = asirdd.map(find)

def findn(x):
    return PNDURA[x]
pnduralist = pnrdd.map(findn)

def finds(x):
    return STEPDURA[x]
stepduralist = steprdd.map(finds)

def findh(x):
    return PHDURA[x]
phduralist = phrdd.map(findh)
```

```python
indexlist = [i for i in range(232744)]
indexrdd = sc.parallelize(indexlist)
```

```python
def duradel(x):
    if pd.isnull(traindata["Opportunity(Default)"][x]):
        return 0
    else:
        deal = traindata["Opportunity(Default)"][x].split("~~")
        init = 0
        for i in deal:
            init = init + int(i)
        return init

oppordd = indexrdd.map(duradel)
```

```python
oppordd.collect()
```

```python
def dura(x):
    if pd.isnull(traindata["Step Duration (sec)"][x]):
        asi = traindata["Anon Student Id"][x]
        pn = traindata["Problem Name"][x]
        step = traindata["Step Name"][x]
        return (ASIDURA[asi] + PNDURA[pn] + STEPDURA[step])/3
    else:
        return traindata["Step Duration (sec)"][x]
    
durardd = indexrdd.map(dura)
durardd.collect()
```

```python
ttraindata = pd.read_csv("data/target_train.csv")

cleardata = pd.DataFrame(indexlist, columns = ["A"])
cleardata.insert(0, "Row", ttraindata["Row"])
del cleardata["A"]

cleardata.insert(1, "Anon Student Id", ttraindata["Anon Student Id"])
cleardata.insert(2, "Problem Hierarchy", ttraindata["Problem Hierarchy"])
cleardata.insert(3, "Problem Name", ttraindata["Problem Name"])
cleardata.insert(4, "Problem View", ttraindata["Problem View"])
cleardata.insert(5, "Step Name", ttraindata["Step Name"])
cleardata.insert(6, "KC(Default)", ttraindata["KC(Default)"])

cleardata.insert(7, "Opportunity(Default)", oppordd.collect())
cleardata.insert(8, "Step Duration (Anon Student Id)", asiduralist.collect())
cleardata.insert(9, "Step Duration (Problem Name)", pnduralist.collect())
cleardata.insert(10, "Step Duration (Step Name)", stepduralist.collect())
cleardata.insert(11, "Step Duration (Problem Hierarchy)", phduralist.collect())

cleardata.insert(12, "Correct First Attempt", ttraindata["Correct First Attempt"])
```
