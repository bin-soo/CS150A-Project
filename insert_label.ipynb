```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
import random

import pyspark as ps
conf = ps.SparkConf()
sc = ps.SparkContext(conf = conf)
```

```python
testpart = pd.read_table("data/test.csv")
testpart
```

```python
indexlist = [i for i in range(1140)]
indexrdd = sc.parallelize(indexlist)
```

```python
results = pd.read_csv("predictions/best results.csv")
```

```python
results
```

```python
origincfa = testpart["Correct First Attempt"]
prediction = results["0"]
```

```python
def replace(x):
    if pd.isnull(origincfa[x]):
        return prediction[x]
    else:
        return origincfa[x]
    
newlist = indexrdd.map(replace)
insertlist = newlist.collect()
insertlist
```

```python
del testpart["Correct First Attempt"]
testpart.insert(13, "Correct First Attempt", insertlist)
testpart
```

```python
testpart.to_csv("test_final.csv")
```
